# PRACTICA No. 3. NEURONA DE McCulloch Pits.
### El modelo de McCulloch-Pitts (MCP) es una simplificación de una neurona, que toma varias entradas, las suma con un peso, y luego aplica una función de activación (generalmente una función escalón) para determinar la salida.
### En esta practica entrenaremos una célula de McCulloch-Pitts para que sea capaz de comportarse como una compuerta lógica NOT.

# Paso 1: Importar librerias

```python
# import numpy as np
```

# Paso 2: Definir la función de activación para el NOT logico

```python
def mcculloch_pitts_NOT(x):
  # Peso de la entrada
  w = -1

  # Umbral
  teta = -0.5

  # Suma ponderada
  wheighted_sum = w * x

  # Salida de la celula
  if wheighted_sum >= teta:
      return 1
  else:
      return 0
```

# Paso 3: Probar la neurona de McCulloch-Pitts para el NOT logico

```python
print(mcculloch_pitts_NOT(0))  # Salida esperada: 1
print(mcculloch_pitts_NOT(1))  # Salida esperada: 0
```
