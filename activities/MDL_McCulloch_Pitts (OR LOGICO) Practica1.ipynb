# PRACTICA No. 1. NEURONA DE McCulloch Pits.
### El modelo de McCulloch-Pitts (MCP) es una simplificación de una neurona, que toma varias entradas, las suma con un peso, y luego aplica una función de activación (generalmente una función escalón) para determinar la salida.
### En esta practica entrenaremos una célula de McCulloch-Pitts para que sea capaz de comportarse como una compuerta lógica OR de dos entradas.

# Paso 1: Importar librerias

```python
# import numpy as np
```

# Paso 2: Definir la función de activación para el OR logico

```python
def mcculloch_pitts_OR(x1, x2):
  # Pesos de las entradas
  w1, w2 = 1, 1

  # Umbral
  teta = 1

  # Suma ponderada de las entradas
  wheighted_sum = w1 * x1 + w2 * x2
  
  # Salida de la celula
  if wheighted_sum >= teta:
      return 1
  else:
      return 0
```

# Paso 3: Probar la neurona de McCulloch-Pitts para el OR logico

```python
print(mcculloch_pitts_OR(0, 0))  # Salida: 0
print(mcculloch_pitts_OR(0, 1))  # Salida: 1
print(mcculloch_pitts_OR(1, 0))  # Salida: 1
print(mcculloch_pitts_OR(1, 1))  # Salida: 1
```
